```python
mobile_name = []
rating = []
price = []
feature = []

for i in range(1, 21):
    url = 'https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
    
    page = requests.get(url)
    html_code = page.text
    
    data = BeautifulSoup(html_code)
    
    # Mobile Names
    mobile_names = data.find_all('div', attrs = {'KzDlHZ'})
    for name in mobile_names:
        mobile_name.append(name.text)
    
    # price
    prices = data.find_all('div', attrs = {'Nx9bqj _4b5DiR'})
    for mobile_price in prices:
        price.append(mobile_price.text)
    
    # rating
    ratings = data.find_all('div', attrs = {'XQDdHH'})
    for mobile_rating in ratings:
        rating.append(mobile_rating.text)
    
    # feature
    features = data.find_all('ul', attrs = {'G4BRas'})
    for mobile_feature in features:
        feature.append(mobile_feature.text)
    
    
    
```
